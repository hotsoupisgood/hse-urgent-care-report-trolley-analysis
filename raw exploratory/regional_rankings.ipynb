```python
import pandas as pd
from pathlib import Path

MODEL_VERSION = "v4.1"
BASE = Path("../data/models")

SCALINGS = {
    "Per 10k":            "wide_weekly_scaledPer10k",
    "Per bed":            "wide_weekly_scaledPerBed",
    "Per budget (B/10k)": "wide_weekly_scaledPerBudgetBillionsPer10k",
    "Per budget (000s)":  "wide_weekly_scaledPerBudgetThousands",
}

frames = {}
for label, stem in SCALINGS.items():
    df = pd.read_csv(BASE / stem / MODEL_VERSION / "ranks.csv", index_col=0)
    frames[label] = df["ranked_mean"]

table = pd.DataFrame(frames)
table.index.name = "Region"
table.index = table.index.str.replace("HSE ", "", regex=False)
table = table.sort_values(by="Per 10k", ascending=True)

table
```
