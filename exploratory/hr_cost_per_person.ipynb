```python
import pandas as pd
```

```python
df_budget   = pd.read_excel("../data/2026_hr_budget.xlsx")
df_pop      = pd.read_csv("../data/encatchment_areas.csv")
```

```python
dict_budget     = df_budget.to_dict(orient='records')[0]
dict_pop        = dict(df_pop.to_records(index=False))
```

```python
dict_budget
```

```python
dict_budget_div_pop = {k: v / dict_pop[k] for k, v in dict_budget.items() if k in dict_pop}
dict_budget_div_pop
```

```python
pd.DataFrame(dict_budget_div_pop.items(), columns=['Region', 'CostPerPerson']).to_csv("../data/hr_cost_per_person.csv", index=False)
```
