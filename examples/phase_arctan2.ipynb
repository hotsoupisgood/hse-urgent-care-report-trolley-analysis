# Phase of a seasonal cycle: 
Comparing these values:
This is what I was measuring:
$${\text{not peak!}} = \frac{52}{2\pi}\arctan2(-\gamma, \beta)      \ \%\ 52$$
But these are correct for finding the peak:
$${\text{peak}} = \frac{52}{2\pi}\arctan2(\gamma, \beta)            \ \%\ 52$$
$${\text{peak}} = (-1) \cdot \frac{52}{2\pi}\arctan2(-\gamma, \beta)\ \%\ 52$$


## Interactive: drag β and γ sliders to see peak move

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

t = np.linspace(0, 52, 500)
omega = 2 * np.pi / 52

def plot_phase(beta=1.0, gamma=0.0):
    y = beta * np.cos(omega * t) + gamma * np.sin(omega * t)
    # compute the with positive gamma
    rad_correct      = np.arctan2(gamma, beta)
    # compute the with negative gamma
    rad_neg_gamma    = np.arctan2(-gamma, beta)
    # this one is correct because arctan2 
    # gives the angle in the correct quadrant, so we just take it as is
    peak_correct     = (52 / (2 * np.pi) * rad_correct) % 52
    # the negative gamma flips the sign of the phase
    peak_incorrect_neg_gamma   = (52 / (2 * np.pi) * rad_neg_gamma) % 52
    # this one is correct because arctan2 gives the angle in 
    # the correct quadrant, so we just take it as is
    peak_correct_neg_gamma   = (-1) * (52 / (2 * np.pi) * rad_neg_gamma)   % 52
    # get the max
    true_peak        = t[np.argmax(y)] % 52 

    A = np.sqrt(beta**2 + gamma**2)

    fig, ax = plt.subplots(3, 1, figsize=(9, 8), dpi=160)
    # ax0, ax1, ax_circle = axes

    # --- Time-domain plot 1 (top) ---
    ax[0].plot(t, y, linewidth=2, label='β cos(ωt) + γ sin(ωt)')
    ax[0].axvline(true_peak+0.5,
                  color='black', linewidth=2,   linestyle='-',  label=f'maximum {true_peak:.1f}')
    ax[0].axvline(peak_correct, color='green', linewidth=1.5, linestyle='--', label=f'(52 / (2 * pi) * arctan2( γ, β) = {rad_correct:+.3f} rad → wk {peak_correct:.1f} ✓')
    ax[0].axvline(peak_incorrect_neg_gamma,   color='red',   linewidth=1.5, linestyle='--', 
                  label=f'(52 / (2 * pi) * arctan2(-γ, β)= {rad_neg_gamma:+.3f} rad → wk {peak_incorrect_neg_gamma:.1f} ✗')
    ax[0].axhline(0, color='grey', linewidth=0.8, linestyle=':')
    ax[0].set_xlabel('Week'); ax[0].set_ylabel('Value')
    ax[0].set_title(f'β={beta:.2f}, γ={gamma:.2f}   |   A={A:.2f}')
    ax[0].legend(loc='upper right', fontsize=8)
    ax[0].set_xlim(-1, 53)

    # --- Time-domain plot 2 (middle) ---
    ax[1].plot(t, y, linewidth=2, label='β cos(ωt) + γ sin(ωt)')
    ax[1].axvline(true_peak+0.5,    color='black', linewidth=2,   linestyle='-',  label=f'maximum {true_peak:.1f}')
    ax[1].axvline(peak_correct-0.5, color='green', linewidth=1.5, linestyle='--',           label=f'(52 / (2 * pi) * arctan2( γ, β) = {rad_correct:+.3f} rad → wk {peak_correct:.1f} ✓')
    ax[1].axvline(peak_correct_neg_gamma,   color='red',   linewidth=1.5, linestyle='--',   label=f'(-1) * (52 / (2 * pi) * arctan2(-γ, β)= {rad_neg_gamma:+.3f} rad → wk {peak_incorrect_neg_gamma:.1f} ✗')
    ax[1].axhline(0, color='grey', linewidth=0.8, linestyle=':')
    ax[1].set_xlabel('Week'); ax[1].set_ylabel('Value')
    ax[1].set_title(f'β={beta:.2f}, γ={gamma:.2f}   |   A={A:.2f}')
    ax[1].legend(loc='upper right', fontsize=8)
    ax[1].set_xlim(-1, 53)

    # --- Unit circle (bottom) ---
    theta_circle = np.linspace(0, 2 * np.pi, 200)
    ax[2].plot(np.cos(theta_circle), np.sin(theta_circle), 'k-', linewidth=0.8, alpha=0.3)
    ax[2].axhline(0, color='grey', linewidth=0.5)
    ax[2].axvline(0, color='grey', linewidth=0.5)

    if A > 1e-9:
        bn, gn = beta / A, gamma / A
    else:
        bn, gn = 1.0, 0.0

    # correct vector: (beta, gamma)
    ax[2].annotate('', xy=(bn, gn), xytext=(0, 0),
                       arrowprops=dict(arrowstyle='->', color='green', lw=2))
    ax[2].plot(bn, gn, 'o', color='green', markersize=8, zorder=5,
                   label=f'(γ, β); {rad_correct:+.2f} rad')

    # sign-flipped vector: (beta, -gamma)
    ax[2].annotate('', xy=(bn, -gn), xytext=(0, 0),
                       arrowprops=dict(arrowstyle='->', color='red', lw=2, linestyle='--'))
    ax[2].plot(bn, -gn, 'x', color='red', markersize=8, zorder=5,
                   label=f'(-γ, β); {rad_neg_gamma:+.2f} rad')

    ax[2].set_xlim(-1.4, 1.4)
    ax[2].set_ylim(-1.4, 1.4)
    ax[2].set_aspect('equal')
    ax[2].set_xlabel('β (cos)')
    ax[2].set_ylabel('γ (sin)')
    ax[2].set_title('Unit circle: phase angle')
    ax[2].legend(loc='upper right', fontsize=8)

    plt.tight_layout()
    plt.show()

interact(
    plot_phase,
    beta    = FloatSlider(value=1.0,  min=-2.0, max=2.0, step=0.05, description='β (cos)'),
    gamma   = FloatSlider(value=0.0, min=-2.0, max=2.0, step=0.05, description='γ (sin)'),
)
```
