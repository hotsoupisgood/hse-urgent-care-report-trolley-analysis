## Total populations

```python
# Load the filtered data back in to verify
df = pd.read_csv('../data/uec_data_2023_2025_full_with_regions.csv')
```

Make Weekday column

```python

df['weekday']= pd.to_datetime(df['Date'], dayfirst=True).dt.day_of_week
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Then sort by date to ensure proper order
df = df.sort_values('Date')
```

Make Encatchment df

```python
# get encatchment and turn into a dict
pop_df=pd.read_csv("../data/encatchment_areas.csv")
pop_dict=dict(zip(pop_df.iloc[:, 0], pop_df.iloc[:, 1]))
# map Region -> population into new column 'Region_Population'
df['Region_Population'] = df['Region'].map(pop_dict)
# quick check of unmapped region values
#print("Unmapped region values:", df[df['Region_Population'].isna()]['Region'].dropna().unique())
```

```python
pop_dict
```

## Weight totals
Rate per 10k people
$$\frac{df_{total\ trollies}}{df_{region\ population}}\cdot 10000$$

```python
df['Trollies Per 10k']=(df['Total']/df['Region_Population'])*10000
```

## Daily 
### Make long df

```python
df_long = df.groupby(['Date','Region'])['Trollies Per 10k'].sum().reset_index()
df_long.columns = ['Date', 'Region', 'Value']
```

### Make wide df

```python
df_wide = df_long.pivot(index='Region', columns='Date', values='Value')
```

### Save

```python
df_wide.to_csv("../data/wide_daily_scaledPer10k.csv")
df_long.to_csv("../data/long_daily_scaledPer10k.csv")
```

# Weekly
### Make long df

```python
df_week_long=df_long.groupby(['Region', pd.Grouper(key='Date', freq='W')]).sum()
```

### Make wide df

```python
df_week_wide=df_week_long.pivot_table(index='Region', columns='Date', values='Value')
```

### Save

```python
df_week_long.to_csv("../data/long_weekly_scaledPer10k.csv")
df_week_wide.to_csv("../data/wide_weekly_scaledPer10k.csv")
```

# Budget-Standardised Rates
Standardise trolley counts by per-10,000 health region budget:
$$\frac{\text{total trollies}}{\text{region budget} / \text{region population}}\cdot 10000$$

```python
budget_df = pd.read_excel('../data/2026_hr_budget.xlsx')
budget_dict = budget_df.iloc[0].to_dict()

# Per-capita budget for each region
per_capita_budget = {region: (budget_dict[region] / pop_dict[region])*10000 for region in budget_dict}
per_capita_budget
```

### Daily

```python
df['Per Capita Budget'] = df['Region'].map(per_capita_budget)
df['Trollies Per Budget/Pop'] = df['Total'] / df['Per Capita Budget']

df_long_budget = df.groupby(['Date', 'Region'])['Trollies Per Budget/Pop'].sum().reset_index()
df_long_budget.columns = ['Date', 'Region', 'Value']

df_wide_budget = df_long_budget.pivot(index='Region', columns='Date', values='Value')

df_long_budget.to_csv('../data/long_daily_scaledPerBudgetPop.csv', index=False)
df_wide_budget.to_csv('../data/wide_daily_scaledPerBudgetPop.csv')
```

### Weekly

```python
df_week_long_budget = df_long_budget.groupby(['Region', pd.Grouper(key='Date', freq='W')]).sum()
df_week_wide_budget = df_week_long_budget.pivot_table(index='Region', columns='Date', values='Value')

df_week_long_budget.to_csv('../data/long_weekly_scaledPerBudgetPop.csv')
df_week_wide_budget.to_csv('../data/wide_weekly_scaledPerBudgetPop.csv')
```
