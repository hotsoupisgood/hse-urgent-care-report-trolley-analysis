```python
import numpy as np
import pandas as pd
from itertools import combinations
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == 'thesis' else Path.cwd()
df = pd.read_csv(ROOT / 'data/models/wide_weekly_scaledPer10k/v2.8/raw_samples.csv')

regions = [
    'HSE Dublin and Midlands', 'HSE Dublin and North East',
    'HSE Dublin and South East', 'HSE Mid West',
    'HSE South West', 'HSE West and North West',
]

def pairwise(period):
    mat = np.column_stack([df[f'delta_{period}[{i+1}]'].values for i in range(6)])
    rows = []
    for a, b in combinations(range(6), 2):
        d = mat[:, a] - mat[:, b]
        if d.mean() < 0:
            d = -d
            a_lbl, b_lbl = regions[b], regions[a]
        else:
            a_lbl, b_lbl = regions[a], regions[b]
        lo, hi = np.quantile(d, [0.025, 0.975])
        rows.append({
            'Comparison': a_lbl,
            'vs': b_lbl,
            'Mean diff': round(float(d.mean()), 2),
            '95% CI': f'[{lo:.2f}, {hi:.2f}]',
            'Contains 0': 'Yes' if lo <= 0 <= hi else 'No',
        })
    return pd.DataFrame(rows).sort_values('Mean diff', ascending=False).reset_index(drop=True)

print('Pairwise comparison: NY+1 (delta_post)')
pairwise('post')
```

```python
print('Pairwise comparison: NY week (delta_mid)')
pairwise('mid')
```

```python
print('Pairwise comparison: NY-1 (delta_pre)')
pairwise('pre')
```
